Now let's join all the DataSet based on columns.

```python
# Assuming BasicInfo_df, BreakthroughInfo_df, and LastWorkInfo_df are your existing DataFrames
# Assuming Imdb_df contains IMDb data
# Merging the data based on common data set for IMDB data

# Merge IMDb data with BasicInfo_df using 'inner' join
Final_data = pd.merge(BasicInfo_df, Imdb_df, left_index=True, right_index=True, how='inner')

# Merge with BreakthroughInfo_df using 'inner' join
Final_data = pd.merge(Final_data, BreakthroughInfo_df, left_index=True, right_index=True, how='inner')

# Merge with LastWorkInfo_df using 'inner' join
Final_data = pd.merge(Final_data, LastWorkInfo_df, left_index=True, right_index=True, how='inner')

# Print or inspect merged_data to verify the merged DataFrame
Final_data
```

```python
# Drop additional columns in the code
Final_data.drop(['Entertainer','Entertainer_y'], axis = 1, inplace = True)
Final_data = Final_data.rename(columns ={'Entertainer_x': 'Entertainer'})
```

After observing the Data set we can say there are 6 missing data in the Year of First Oscar/Grammy/Emmy, 40 missing data in the column Year of Death, 5 missing data from the column Meta_score and 7 from Gross column.

Firstly converting the decimal type number into proper interger fomat, from the given columns Year of First Oscar/ Grammy / Emmy awards and Year of Death.

```python
#Convert columns to integers
Final_data['Year of First Oscar/Grammy/Emmy'] = Final_data['Year of First Oscar/Grammy/Emmy'].astype(float).astype('Int64')
Final_data['Year of Death'] = Final_data['Year of Death'].astype(float).astype('Int64')
```

Replacing the Null / Missing values with some content

```python
Final_data['Year of First Oscar/Grammy/Emmy'] = Final_data['Year of First Oscar/Grammy/Emmy'].astype('object')
Final_data['Year of Death'] = Final_data['Year of Death'].astype('object')

#Filling the missing/null data with some content.
Final_data.fillna({'Year of First Oscar/Grammy/Emmy': 'Not Available', 'Year of Death': 'Alive', 'Meta_score': 'Not available', 'Gross': 0}, inplace=True)
```

Creating lists to store the extracted data of Persons like number of films act, height, quotes, pics from IMDB data set.

```python
pics = []
num_films_act = []
quotes = []
height = []

for i in range(len(BasicInfo_df)):
    ent_name = BasicInfo_df.Entertainer[i]
    print("Starting-->", ent_name)
    people = mi.search_person(ent_name)
    if len(people):
        try:
            per = mi.get_person(people[0].personID)
        except:
            per = people[0]

        keys = per.keys()

        #Picture of the person
        if "full-size headshot" in keys:
            pics.append(per["full-size headshot"])
        elif "headshot" in keys:
            pics.append(per["headshot"])
        else:
            pics.append("NA")

        #count of flims acted
        if "filmography" in keys:
            if "actor" in per["filmography"]:
                num_films_act.append(len(per["filmography"]["actor"]))
            elif "actress" in per["filmography"]:
                num_films_act.append(len(per["filmography"]["actress"]))
            else:
                num_films_act.append(0)
        else:
            num_films_act.append(0)

        #height
        if "height" in keys:
            height.append(per["height"])
        else:
            height.append("NA")

        #Quotes
        if "quotes" in keys:
            quotes.append(per["quotes"][0])
        else:
            quotes.append("NA")
    else:
        height.append("Np")
        pics.append("Np")
        num_films_act.append("Np")
        quotes.append("Np")

    print("Done----->", ent_name)
    print("    ")
```

```python
#Checking the data length
print(len(pics))
print(len(height))
print(len(num_films_act))
print(len(quotes))
```

```python
#Adding the extracted data to the final data set
Final_data["Picture"] = pics
Final_data["Height"] = height
Final_data["Number_films_act"] = num_films_act
Final_data["Quote"] = quotes
```

Create a different CSV file for actor and actress names, movies names and publish year of movies for all the actor and actress.

```python
#List for actor names, movies and publish year of the movies
actor = []
movie_title = []
year = []

#Function to get all the actor and actress name, movies done and movie publish year from IMDB
for i in range(len(BasicInfo_df)):
    ent_name = BasicInfo_df.Entertainer[i]
    leen = len(movie_title)
    print("Start-->", ent_name)
    persons = mi.search_person(ent_name)
    try:
        per = mi.get_person(persons[0].personID)
        if 'filmography' in per.keys():
            ma_keys = per["filmography"].keys()
            if "actor" in ma_keys:
                act = per["filmography"]["actor"]

                for j in range(len(act)):
                    movi = mi.get_movie(act[j].movieID)
                    k = movi.keys()
                    actor.append(ent_name)
                    if "title" in k:
                        movie_title.append(movi["title"])
                    else:
                        movie_title.append("NA")

                    if "year" in k:
                        year.append(movi["year"])
                    else:
                        year.append("NA")

            elif "actress" in  ma_keys:
                act = per["filmography"]["actress"]

                for j in range(len(act)):
                    movi = mi.get_movie(act[j].movieID)
                    k = movi.keys()
                    actor.append(ent_name)
                    if "title" in k:
                        movie_title.append(movi["title"])
                    else:
                        movie_title.append("NA")

                    if "year" in k:
                        year.append(movi["year"])
                    else:
                        year.append("NA")
            else:
                actor.append(ent_name)
                movie_title.append("NA")
                year.append("NA")
        else:
            actor.append(ent_name)
            movie_title.append("NA")
            year.append("NA")
    except:
        actor.append(ent_name)
        movie_title.append("NA")
        year.append("NA")
    print("Done-->", ent_name, " Added--> ", len(movie_title)-leen, " Movies")
    print("   ")
```

```python
#Checking the data length
print(len(actor))
print(len(movie_title))
print(len(year))
```

```python
#Convert the list into a proper Dataframe format
Actors_films_list_df = pd.DataFrame({"Actor Name":actor, "Movie Title":movie_title, " Year of Publish":year})
```
